```python
import os,json,datetime
path = 'cdlog'

# SELECT payload.* FROM cdlog WHERE payload.received BETWEEN '2018-07-18T03:00:00' AND '2018-07-18T10:00:00'
start_dt = datetime.datetime.strptime('2018-07-18T03:00:00',"%Y-%m-%dT%H:%M:%S")
end_dt = datetime.datetime.strptime('2018-07-18T10:00:00',"%Y-%m-%dT%H:%M:%S")

filtered_data = []

for file in os.listdir(path):
    print (file) 
    fp = os.path.join(path,file)
    with open(fp, 'r') as f:
        for line in f:
            data = json.loads(line)
            payload = json.loads(data['payload'])
            #print(payload['received'])
            rdo = None
            try:
                rdo = datetime.datetime.strptime(payload['received'],"%Y-%m-%dT%H:%M:%S.%f")
            except Exception as e:
                print(e)
            if rdo is not None:
                if rdo > start_dt and rdo < end_dt:
                    filtered_data.append(payload)        
```
