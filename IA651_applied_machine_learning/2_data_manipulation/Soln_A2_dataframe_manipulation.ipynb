```python
#Assignment 

import pandas as pd 
import datetime
import warnings
```

```python
#- load the northeast_realestate.parquet
df = pd.read_parquet('northeast_realestate.parquet')
df
```

```python
#- Replace the zipcode with a zero padded string, head(10) the result
df['zip_code'] = df['zip_code'].astype(str).str.split('.').str[0]
df['zip_code'] = df['zip_code'].str.zfill(5)
df.head(10)
```

```python
#- Find unique properties in Rye, NY.  head(15) the result
df2 = df[(df['city'] == 'Rye') & (df['state'] == 'New York')]
df2.head(15)
```

```python
#- Show only properties sold in March and April. head(5) the result

df2['month'] = pd.DatetimeIndex(df2['sold_date']).month
warnings.filterwarnings("ignore")

df3 = df2[(df2['month'] == 3.0) | (df2['month'] == 4.0)]
df3.head(10)
```

```python
#- What is the formatting string of the sold_date?
print(type(df3['sold_date']))
#It is class object which returns month, year and day of the month. mm-dd-yyyy
```

```python
#- What is the difference between the average price of a home sold with 
#St in the address (on a street) and Rd in the address (on a road)

# df4 = df.loc[df['street'].str.contains('St').any(), 'price']
df4 = df[df['street'].str.contains('St ', na=False)]
df5 = df[df['street'].str.contains('Rd ', na=False)]

a = df4['price'].mean()
print("Avg. St. House Price: ", a)

b = df5['price'].mean()
print("Avg. Rd. House Price: ", b)
print('Difference: Avg. St price - Avg. Rd price', a-b)

#average price of St. houses is greater than Rd. mentioned houses. 
```

```python
#- Which are the top 20 cities with highest home sale prices in NY? (sorted by price DESC)
df6 = df[df['state'] == "New York"]
df6.sort_values('price', ascending = False).head(20)
#df6.head(20)
```
